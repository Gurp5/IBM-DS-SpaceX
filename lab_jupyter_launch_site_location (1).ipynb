```python
# Task 2: Mark the success/failed launches for each site on the map
# Save the map as an HTML file
site_map.save("spacex_launch_map.html")
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]

# Create a map centered on a general location (you can adjust as needed)
site_map = folium.Map(location=[28.562302, -80.577356], zoom_start=5)

# Iterate over each row in spacex_df and add markers to the map
for index, row in spacex_df.iterrows():
    # Determine the color based on success (1) or failure (0)
    launch_color = 'green' if row['class'] == 1 else 'red'
    
    # Add a marker for each launch
    folium.Marker(
        location=[row['Lat'], row['Long']],
        popup=f"Launch Site: {row['Launch Site']}\nOutcome: {'Success' if row['class'] == 1 else 'Failure'}",
        icon=folium.Icon(color=launch_color, icon='rocket', prefix='fa')
    ).add_to(site_map)

# Save the map as an HTML file
site_map.save("spacex_launch_map.html")

# Display the map
site_map
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python

launch_site_lat = 28.562302  # Launch site latitude
launch_site_lon = -80.577356  # Launch site longitude

coastline_lat = 28.56367  # Closest coastline latitude
coastline_lon = -80.57163  # Closest coastline longitude

# Calculate the distance between the launch site and the coastline
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Output the distance
print(f"Distance between launch site and coastline: {distance_coastline:.2f} km")
```

```python
from folium import DivIcon

# Create a marker for the closest coastline point
coastline_coords = [coastline_lat, coastline_lon]

# Create a DivIcon to display the distance
distance_marker = folium.Marker(
    location=coastline_coords,
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_coastline),
    )
)

# Add the marker to the map
site_map.add_child(distance_marker)

# Display the map
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
# Define the coordinates of the launch site and the closest city, railway, or highway
site_map.save("spacex_launch_map.html")
launch_site_coords = [28.56367, -80.57163]  # Launch site coordinates
city_coords = [28.5259, -81.3813]  # Example: coordinates of a city (e.g., Orlando)

# Calculate the distance between the launch site and the city
distance_to_city = calculate_distance(launch_site_coords[0], launch_site_coords[1], city_coords[0], city_coords[1])

# Create a PolyLine object to draw the line between the launch site and the city
line_to_city = folium.PolyLine(
    locations=[launch_site_coords, city_coords],
    weight=2,  # Line thickness
    color='green',  # Line color
    opacity=0.7  # Line transparency
)

# Create a marker for the city with the distance information as a popup
city_marker = folium.Marker(
    location=city_coords,
    popup=f"City: Distance to launch site: {distance_to_city:.2f} km",
    icon=folium.Icon(color='blue', icon='info-sign')
)

# Add the PolyLine and Marker to the map
site_map.add_child(line_to_city)
site_map.add_child(city_marker)

# Display the map
site_map
```
